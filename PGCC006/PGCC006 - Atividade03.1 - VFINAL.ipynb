```
'''
Dado um grafo (em anexo), implemente os seguintes algoritmos gulosos para obter a árvore de custo mínima: 
1) Prim’s e 
2) Kruskal’s. 

Todos os algoritmos devem ser implementados em Python. 
Utilize o código fornecido pelo professor (em anexo) para testar os algoritmos. 
Entretanto, faça outros testes caso considere necessário! 

Lembrando que não é possível saber se um algoritmo está correto, apenas fazendo testes. 
Os testes só ajudam a minimizar erros, mas não garantem que eles não existem!!! 

Você deve enviar o código dos dois algoritmos.
'''
import math
from heapq import heappush, heappop

class Graph:
  def __init__(self):
    self.vertexMap = dict()           

  def addVertex(self, v):
    self.vertexMap[v] = dict()

  def removeVertex(self, v):  
    if v in self.vertexMap:
      for (i,j) in self.vertexMap[v].copy():
        print(f"e->{(i,j)}")
        self.removeEdge(i,j)
      del self.vertexMap[v]

  def vertices(self):    
    return list(self.vertexMap.keys())

  def adjacents(self, v):
    return [j for (i, j) in self.outgoing(v)]   

  def addEdge(self, u,v,data):
    if (u in self.vertexMap) and (v in self.vertexMap):
      self.vertexMap[u][(u,v)] = data
      self.vertexMap[v][(v,u)] = data
    else:
      raise ValueError(f"One or both of the V {u} and {v} are not present in the Graph!")  

  def removeEdge(self,u,v):
    if ((u,v) in self.vertexMap[u]) and ((v,u) in self.vertexMap[v]):
      del self.vertexMap[u][(u,v)]      
      del self.vertexMap[v][(v,u)]  

  def edges(self):
    ret = []
    for e in self.vertexMap.values():
      if len(e.keys()):
        ret.extend(list(e.keys()))
    return ret

  def getEdge(self,u,v):
    return self.vertexMap[u][(u,v)]

  def outgoing(self, v):
    return list(self.vertexMap[v].keys())  

  def outdegree(self, v):
    return len(self.vertexMap[v])

  def incoming(self, v):
    return [(j,i) for (i,j) in self.vertexMap[v]] 

  def indegree(self, v):
    return len(self.vertexMap[v])

  def path(self, v):
    ret = ""
    visited = set()
    visited.add(v)
    stack = []
    stack.append((v,None))
    while stack:
      (v, p) = stack.pop()
      if p:        
        ret+=f"{p}--{self.getEdge(p,v)}--{v}  "
      for u in self.adjacents(v):        
        if u not in visited:          
          visited.add(u)          
          stack.append((u,v))
    return ret.strip() 
# ---------------------------------------------------------------------------
### BFS ########################################################################
def bfs(G,v):                                                 
    L = list()                                                
    Q = list()                                                
    S = set()                                                 
    Q.append(v)                                               
    S.add(v)                                                  
    while Q != list():                                        
        e = Q.pop(0)                                          
        for u in G.adjacents(e):                              
            if u not in S:                                    
                S.add(u)                                      
                Q.append(u)                                   
                L.append(u)                                   
    return L 
### KRUSKAL ####################################################################
#To do
def kruskals(G):                                                                # recebo um grafo
    M = Graph()                                                                 # crio variável para construir um 2o grafo do tipo graph
    S = set()                                                                   # crio variável para receber arestas visitados e sem gerar ciclo num set
    L = []                                                                      # crio variável para receber as arestas numa lista
    for e in G.edges():                                                         # percorro as arestas
        L.append((G.getEdge(e[0], e[1]), e[0], e[1]))                            # adiciono na lista de arestas a distancia, origem, destino
    L.sort()                                                                    # ordeno a lista
    t = len(G.vertices())                                                       # pego o quantidade dos vertices em G
    for a in L:                                                                 # percorro lista ordenada de aresta...
        if ((a[1], a[2]) not in M.edges()):                                     # se a tupla ( aresta ) não estiver no grafo novo...
            if a[1] in M.vertices():                                            # se o primeiro vertice existir no conjunto de vertices do grafo novo...
                path = M.path(a[1]).split()                                     # path = recebe o caminho percorrido do primeiro vertice
                #path = bfs(M,a[1])
                for p in path:                                                  # percorro o caminho do vertice ...
                    if (p[0] == a[2] or p[-1] == a[2]):                         # verifico se os vertices (a[2]) existe e está interligados tanto na ida quanto na volta
                        S.add((a[1], a[2]))                                     # adciono a aresta ao set na ida...
        if ((a[1], a[2]) not in S):                                             # se não formar clico, adciono a aresta ao novo grafo
            if a[1] not in M.vertices():                                        # se aresta 1 não estiver no conjunto de vertices...
                M.addVertex(a[1])                                               # adciono vertice 1 ao novo grafo.
            if a[2] not in M.vertices():                                        # se aresta 2 não estiver no conjunto de vertices...
                M.addVertex(a[2])                                               # adciono vertice 2 ao novo grafo.
            M.addEdge(a[1], a[2],a[0])                                          # adciono aresta e a distancia ao grafo novo
    return M                                                                    # retona a minimal spanning tree ( árvore geradora mínima )

### PRIMS ######################################################################
#To do
def prims(G):                                                                   # recebo um grafo
    M = Graph()                                                                 # crio variavel para construir um 2o grafo do tipo graph
    D = math.inf                                                                # crio variável para receber a distancia entre os vertices iniciando com infinito
    for i in G.edges():                                                         # percorro as arestas de G
        if G.getEdge(i[0], i[1]) < D:                                           # se a distância da aresta for menor que D...
            v1 = i[0]                                                           # adciono 1o vertice ligado ao 2o
            v2 = i[1]                                                           # adciono 2o vertice ligado ao 1o    
            D = G.getEdge(v1, v2)                                               # pego a distancia entre os vertices
    M.addVertex(v1)                                                             # adciono o vertice 0 no grafo novo
    M.addVertex(v2)                                                             # adciono o vertice 1 no grafo novo   
    M.addEdge(v1,v2, D)                                                         # adiciono a aresta e a distancia no grafo novo 
    t = len(G.vertices())                                                       # pego o tamanho dos vertices em G
    while (len(M.edges())//2 < t-1):                                            # enquanto (|E|<|V|-1) adciono a aresta com menor custo na árvore, desde que, ao adcionar não gere ciclo.
        a = 0                                                                   # crio variável vazia para receber o vertice adjacente de menor distancia
        d = math.inf                                                            # crio variável para receber a distancia entre os vertices
        for e in M.vertices():                                                  # percorro o vertices no grafo
            for u in G.adjacents(e):                                            # pego o adjacente vindo do grafo antigo
                if(G.getEdge(e, u) < d) and u not in M.vertices():              # se o distancia entre o vertice do grafo novo e o adjacente vindo do grafo antigo for menor que a menor distância e o adjacente 'u' não estiver no grafo novo...
                    v = e                                                       # 'vertice' é marcado como visitado
                    a = u                                                       # o menor vertice adjacente é atualizado
                    d = G.getEdge(e, u)                                         # a menor distancia recebe a aresta de menor distancia 
        M.addVertex(a)                                                          # adiciono o próximo vertice adjacente
        M.addEdge(v, a, d)                                                      # adiciono a aresta de menor distancia entre os vertices
    return M                                                                    # retorno o grafo novo com menor caminho   (Minimal Spanning Tree)
```
